```python
def wasrab(n, k):
    if n < 2: return n
    return wasrab(n - 1, k) + k*wasrab(n - 2, k)

wasrab(29, 2)
```

```python
def read_fasta(infile):
    fin = open(infile, "r")
    seq = ""
    header = None
    all_seq = {}
    for line in fin:
        if line.startswith(">"):
            if header is not None:
                all_seq[header] = seq
            header = line.lstrip(">").rstrip("\n")
            seq=""
        else:
            seq += line.strip()
    all_seq[header] = seq     
    fin.close()
    return(all_seq)

def rankGC(infile):
    all_seq = read_fasta(infile)
    countGC = {}
    for header, seq in all_seq.items():
        Gcount = 0
        Ccount = 0
        for c in seq:
            if c == 'G':
                Gcount += 1
            elif c == 'C':
                Ccount += 1
        GCpercentage = (Gcount + Ccount) / len(seq) *100
        countGC[header] = GCpercentage
    max_seq = max(countGC, key=countGC.get)
    max_val = countGC[max_seq]
    return max_seq, max_val

infile = "rosalind_gc.txt"
res = rankGC(infile)
print (res)
```
