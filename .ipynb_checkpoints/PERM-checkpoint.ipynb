```python
#Ihsan Muchsin

import itertools as it

def perm(infile, outfile):
    '''
    Given: 1) An input file containing a positive integer nâ‰¤7. 2) The name of output file
    Return: The total number of permutations of length n, followed by a list of all such permutations (in any order). Plus an output file containing the result.
    '''
    
    with open(infile, 'r') as fin:
        n = int(fin.readline())
    
    num_list = [i for i in range(1, n+1)]
    possible_perm = list(it.permutations(num_list)) #list of permutation
    total_perm = len(possible_perm) #total number of permutation
    
    with open(outfile, 'w') as fout:
        fout.write(str(total_perm) + '\n')
        
        possible_perm_str = []
        for el in possible_perm:
            e_str = [str(e) for e in el]
            el_str = ' '.join(e_str)
            possible_perm_str.append(el_str) #convert to string so it is possible to write in output
            
        fout.write('\n'.join(possible_perm_str))
    
    return (total_perm, possible_perm)
```

```python
infile = 'rosalind_perm.txt'
outfile = 'perm_sol.txt'

res = perm(infile, outfile)

print(res[0])
for p in res[1]:
    print(*p)
```
